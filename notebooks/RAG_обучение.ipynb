```python
class Vectorizer(nn.Module):
    def __init__(self, bert: SentenceTransformer, hid_size=1536, output_size=1536, dropout=0.1):
        """
        :param bert: frozen encoder (FRIDA)
        :param hid_size: размер скрытых слоёв MLP
        :param output_size: размер выходного embedding (должен совпадать с embedding FRIDA)
        :param dropout: dropout rate
        """
        super().__init__()
        self.bert = bert
        # Заморозим BERT
        for param in self.bert.parameters():
            param.requires_grad = False

        # Верхняя голова: 2 слоя MLP с ReLU и Dropout
        self.mlp = nn.Sequential(
            nn.Linear(output_size, hid_size),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hid_size, output_size)  # линейный слой для векторного пространства FRIDA
        )

    def forward(self, input_texts):
        """
        :param input_texts: список строк [batch_size]
        :return: [batch_size, output_size]
        """
        # 1. Получаем embedding из frozen FRIDA
        with torch.no_grad():
            embeddings = self.bert.encode(input_texts, convert_to_tensor=True, normalize_embeddings=True)

        # 2. Прогоняем через MLP (dropout активен только на train)
        out = self.mlp(embeddings)
        # Нормализуем для косинусной близости
        out = F.normalize(out, dim=-1)
        return out
```

```python
class MLP_head(nn.Module):
    def __init__(self, hid_size=1536, output_size=1536, dropout=0.1):
        """
        :param hid_size: размер скрытых слоёв MLP
        :param output_size: размер выходного embedding (должен совпадать с embedding FRIDA)
        :param dropout: dropout rate
        """
        super().__init__()

        self.mlp = nn.Sequential(
            nn.Linear(output_size, hid_size),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hid_size, output_size)
        )

    def forward(self, embeddings):
        """
        :param emd_after_FRIDA: [batch_size, embeddings]
        :return: [batch_size, output_size]
        """

        out = self.mlp(embeddings)
        # Нормализуем для косинусной близости
        out = F.normalize(out, dim=-1)
        return out
```

```python
# Сначала определим правильную функцию encode_texts
def encode_texts(texts, model, prompt_name, batch_size=256, device="cuda"):
    """
    Правильная функция для кодирования текстов с SentenceTransformer
    """
    embeddings = []

    # Переносим модель на нужное устройство
    model = model.to(device)

    with torch.inference_mode(), torch.amp.autocast("cuda", dtype=torch.bfloat16):
        for i in tqdm(range(0, len(texts), batch_size)):
            batch = texts[i:i + batch_size]
            # Правильный вызов: используем метод encode, а не прямой вызов модели
            batch_emb = model.encode(
                batch,
                convert_to_tensor=True,
                batch_size=batch_size,
                device=device,
                normalize_embeddings=True,  # Нормализуем сразу в encode
                prompt_name=prompt_name
            )
            embeddings.append(batch_emb.cpu())

    return torch.cat(embeddings, dim=0)

```

```python
def correct_evaluation_for_colisions_articles(data):
    '''
    Линейный проход по сплиту Dataset с формированием 3 списков: вопросов, документов,
    индексов док<->вопрос
    с учётом коллизии между вопросами и документами
    Должна правильно считать отображение статья<->вопрос
    '''
    query = []
    articles = []
    idx = []
    idx_article_query=[]
    tuple_for_current_article = []

    current_doc_idx = data[0]['doc_id']
    current_dic = 0
    query.append(data[0]['question'])
    articles.append(data[0]['text'])
    idx.append(current_dic)
    tuple_for_current_article.append(0)

    for i in range(1,len(data)):
        example = data[i]
        if example['doc_id'] == current_doc_idx:
            tuple_for_current_article.append(i)
            query.append(example['question'])
            idx.append(current_dic)
        else:
            idx_article_query.append(tuple(tuple_for_current_article))
            query.append(example['question'])
            articles.append(example['text'])
            current_dic += 1
            current_doc_idx = example['doc_id']
            idx.append(current_dic)
            tuple_for_current_article = [i]

    idx_article_query.append(tuple(tuple_for_current_article))

    return query, articles, idx, idx_article_query
```

```python
def recall_at_k(
    query_embeddings: np.ndarray,  # [N, d]
    doc_embeddings: np.ndarray,    # [M, d]
    gt_doc_ids: list[int],         # len N
    ks=(1, 5, 10)
):
    """
    ПРОБЛЕМА: ДОКУМЕНТЫ ПОВТОРЯЮТСЯ!!! Нужно отсеять лишние документы!
    Корректная offline-валидация bi-encoder retriever.
    Один релевантный документ на запрос.
    gt_doc_ids - это иденксы соответствующих статей для вопросов, так как у нас
    в датасете вопрос-ответ, то индекс статьи это просто индекс вопроса,
    gt_doc_ids = [x for x in range(N)]
    """

    assert query_embeddings.shape[1] == doc_embeddings.shape[1]
    assert len(query_embeddings) == len(gt_doc_ids)

    recalls = {k: 0 for k in ks}

    # cosine similarity (при условии L2-normalized эмбеддингов)
    sims = query_embeddings @ doc_embeddings.T  # [N, M]

    for i in range(len(query_embeddings)):
        ranking = np.argsort(-sims[i])  # индексы документов по убыванию
        gt_id = gt_doc_ids[i]

        for k in ks:
            if gt_id in ranking[:k]:
                recalls[k] += 1

    return {k: recalls[k] / len(query_embeddings) for k in ks}
```

```python
metrics = recall_at_k(
    query_embeddings=query_embs,
    doc_embeddings=doc_embs,
    gt_doc_ids=gt_ids,
    ks=(1, 5, 10)
)

for k, v in metrics.items():
    print(f"Recall@{k}: {v:.4f}")
```

```python
def prepare_for_recall_at_k(data):
    """
    Готовит данные для расчёта recall@k без зависимости от порядка.
    Возвращает:
        queries: list[str] - все вопросы
        unique_articles: list[str] - уникальные статьи
        relevant_indices: list[int] - индексы правильных статей для каждого вопроса
    """
    # Словари для маппинга
    doc_id_to_idx = {}  # doc_id -> индекс в unique_articles
    unique_articles = []  # список уникальных текстов статей

    queries = []
    relevant_indices = []

    # Собираем уникальные статьи и маппинг
    for example in data:
        doc_id = example['doc_id']
        if doc_id not in doc_id_to_idx:
            doc_id_to_idx[doc_id] = len(unique_articles)
            unique_articles.append(example['text'])

    # Формируем списки вопросов и индексы правильных статей
    for example in data:
        queries.append(example['question'])
        relevant_indices.append(doc_id_to_idx[example['doc_id']])

    return queries, unique_articles, relevant_indices

def prepare_with_article_groups(data):
    '''
    Основная функция, она вызвает ту что сверху, нужна для реализации обоих типов обучения: через вопросы и через статьи
    '''
    queries, unique_articles, relevant_indices = prepare_for_recall_at_k(data)

    # Группируем индексы вопросов по статьям
    article_to_queries = {i: [] for i in range(len(unique_articles))}
    for q_idx, doc_idx in enumerate(relevant_indices):
        article_to_queries[doc_idx].append(q_idx)

    # Сортируем для воспроизводимости
    idx_article_query = [
        tuple(sorted(queries_list))
        for doc_idx in range(len(unique_articles))
        for queries_list in [article_to_queries[doc_idx]]
    ]

    return queries, unique_articles, relevant_indices, idx_article_query
```

```python
def info_nce_loss(query_embeddings, passage_embeddings, temperature=0.05):
    """
    Это функцинальная реализация MultipleNegativesRankingLoss (InfoNCE)

    """
    # 1. Нормализуем эмбеддинги для косинусного сходства
    query_embeddings = F.normalize(query_embeddings, p=2, dim=1)
    passage_embeddings = F.normalize(passage_embeddings, p=2, dim=1)

    # 2. Вычисляем матрицу сходства (batch_size x batch_size)
    # scores[i][j] - сходство i-го запроса с j-м документом
    logits = torch.matmul(query_embeddings, passage_embeddings.transpose(0, 1)) / temperature

    # 3. Создаем таргеты: на диагонали должны быть правильные пары (0, 1, 2, ..., N)
    labels = torch.arange(query_embeddings.size(0), device=query_embeddings.device)

    # 4. Считаем кросс-энтропию
    loss = F.cross_entropy(logits, labels)
    return loss
```

```python
def get_bm25_negatives(
    question: str,
    pos_doc_id: int,# который положительный к вопросу
    k: int = 15
):
    '''
    Возвращает индексы bm25 negatives
    '''
    # Токенизируем запрос
    tokenized_q = word_tokenize(question.lower())


    scores = np.array(bm25.get_scores(tokenized_q))

    # 3. Находим индексы Top-K  не сортируя весь массив
    # Нам нужно k+1, так как один из них может быть положительным документом
    k_to_find = k + 1
    if k_to_find > len(scores): k_to_find = len(scores)

    # argpartition перемещает n самых больших значений в конец массива
    # Индексы будут не отсортированы, но это будут именно ТОП-N индексов
    top_indices = np.argpartition(scores, -k_to_find)[-k_to_find:]

    # 4. Теперь сортируем ТОЛЬКО эти найденные индексы (это быстро, их всего 16-100)
    ranked = top_indices[np.argsort(-scores[top_indices])]


    hard_negs = [
        i for i in ranked
        if i != pos_doc_id
    ][:k]

    return hard_negs
```

```python
from pyserini.index import IndexReader
from pyserini.search import LuceneSearcher
import numpy as np
from tqdm import tqdm
import json
import os

#  Подготовка данных
def prepare_pyserini_index(documents, index_dir="bm25_index"):
    os.makedirs(index_dir, exist_ok=True)

    # Сохраняем документы в формате JSONL
    with open(f"{index_dir}/docs.jsonl", "w") as f:
        for i, doc in enumerate(documents):
            json.dump({"id": str(i), "contents": doc}, f)
            f.write("\n")

    # Создаем индекс
    os.system(f"python -m pyserini.index.lucene \
              --collection JsonCollection \
              --input {index_dir} \
              --index {index_dir} \
              --generator DefaultLuceneDocumentGenerator \
              --threads 8")

#  Быстрая функция майнинга
def get_bm25_negatives_batch(questions, pos_doc_ids, k=15):
    """
    Пакетная обработка всех запросов за 1 проход

    Args:
        questions: List[str] - все вопросы
        pos_doc_ids: List[int] - ID позитивных документов для каждого вопроса
        k: int - количество негативов
    Returns:
        List[List[int]] - негативы для каждого вопроса
    """
    searcher = LuceneSearcher("bm25_index")
    searcher.set_bm25(k1=0.9, b=0.4)

    results = []

    for i, (question, pos_id) in enumerate(zip(tqdm(questions), pos_doc_ids)):
        # Выполняем поиск для текущего вопроса
        hits = searcher.search(question, k=k+10)  # Берем с запасом

        # Фильтруем позитивный документ и берем топ-k
        negs = []
        for hit in hits:
            doc_id = int(hit.docid)
            if doc_id != pos_id and len(negs) < k:
                negs.append(doc_id)
            if len(negs) == k:
                break

        # Если не хватает негативов, заполняем случайными
        while len(negs) < k:
            rand_id = np.random.randint(0, len(searcher.index_reader.stats()["documents"]))
            if rand_id != pos_id and rand_id not in negs:
                negs.append(rand_id)

        results.append(negs)

    return results


# Сначала создаем индекс
prepare_pyserini_index(articles_train, index_dir="bm25_index")# вот сюда добавить отфильтрованные тексты!

# Затем майнинг за минуты ЗДЕСЬ НУЖНО ПЕРЕДАВАТЬ НЕ СЫРЫЕ ДАННЫЕ !!!!!!!!
bm25_negatives = get_bm25_negatives_batch(
    questions=query_train,
    pos_doc_ids=query_to_article_train,
    k=50
)
```

```python
class MyDataset_via_articles_with_BM25_negatives(Dataset):
    def __init__(self, query, articles, article_idx, id_query_to_negatives):
        """
        На вход уже подавать тензоры: [число вопросов/статей, эмбединг]
        article_idx: List[tuples]
        """
        self.query = query
        self.articles = articles
        self.article_idx = article_idx
        self.id_query_to_negatives = id_query_to_negatives

    def __len__(self):
        """
        Должен возвращать общее количество вопросов в датасете.
        """
        return self.articles.shape[0]

    def __getitem__(self, idx):
        """
        Логика получения одного примера по индексу idx.
        Берём эмбединг документа и к нему по индексу случайно берём один эмбединг вопроса
        Помжет для избегания false negatives
        """
        index_query = random.choice(self.article_idx[idx])
        query = self.query[index_query]

        negatives = self.articles[self.id_query_to_negatives[index_query]]# берём из тензора негативы по индексу
        return {
            'query': query,
            'article': self.articles[idx],
            'bm25_negatives': negatives
        }

# Определяем модель для обучения с негативами:
class BiEncoder_MLP_BM25_negatives(nn.Module):
    def __init__(self, query_model, doc_model):
        super().__init__()
        self.query_encoder = query_model
        self.doc_encoder = doc_model
    @property
    def device(self):
        # Берем устройство, на котором лежит первый параметр модели
        return next(self.parameters()).device

    def forward(self, batch):
        # batch = {'query': (batct_dim, tensor_embedding), 'article': (batct_dim, tensor_embedding), 'bm25_negatives':(tensor)}
        if self.training:
            q_emb = self.query_encoder(batch['query'])
            d_emb = self.query_encoder(batch['article'])
            negatives =  self.query_encoder(batch['bm25_negatives'])
            return q_emb, d_emb, negatives
        else:
            q_emb = self.query_encoder(batch['query'])
            d_emb = self.query_encoder(batch['article'])
            return q_emb, d_emb

def info_nce_loss_BM25_negative(query_embeddings, passage_embeddings, bm25negatives, temperature=0.005): # Экстремально низкое значение!
    """
    Это функциональная реализация MultipleNegativesRankingLoss (InfoNCE) с негативами MB25
    ВАЖНО: passage_embeddings - это именно позитивные эмбединги!!!
    """
    #  Нормализуем эмбеддинги для косинусного сходства
    query_embeddings = F.normalize(query_embeddings, p=2, dim=1)
    passage_embeddings = F.normalize(passage_embeddings, p=2, dim=1)
    bm25negatives = F.normalize(bm25negatives, p=2, dim=1)
    #  Вычисляем матрицу сходства (batch_size x batch_size)
    # scores[i][j] - сходство i-го запроса с j-м документом
    #logits = torch.matmul(query_embeddings, passage_embeddings.transpose(0, 1)) / temperature
    pos_scores = (query_embeddings * passage_embeddings).sum(-1)        # [B]
    neg_scores = torch.einsum("bd,bkd->bk", query_embeddings, bm25negatives)  # [B, K]
    logits = torch.cat([pos_scores.unsqueeze(1), neg_scores],dim=1) /temperature # [B, 1+K]
    #  Создаем таргеты: на диагонали должны быть правильные пары (0, 1, 2, ..., N)
    #labels = torch.arange(query_embeddings.size(0), device=query_embeddings.device)
    labels = torch.zeros(query_embeddings.shape[0], dtype=torch.long, device=query_embeddings.device)
    #  Считаем кросс-энтропию
    loss = F.cross_entropy(logits, labels)
    return loss
```

```python
import torch
import torch.nn.functional as F


def info_nce_loss_with_inbatch_and_bm25(
    query_embeddings: torch.Tensor,      # [B, D]
    passage_embeddings: torch.Tensor,    # [B, D] (positives)
    bm25negatives: torch.Tensor,          # [B, K, D]
    temperature: float = 0.04
):
    """
    Полноценный InfoNCE:
    - in-batch negatives
    - BM25 hard negatives
    """

    B, D = query_embeddings.shape
    K = bm25negatives.shape[1]

    #  L2 normalization
    query_embeddings = F.normalize(query_embeddings, dim=1)
    passage_embeddings = F.normalize(passage_embeddings, dim=1)
    bm25negatives = F.normalize(bm25negatives, dim=2)

    #  in batch similarity: [B, B]
    inbatch_logits = torch.matmul(
        query_embeddings,
        passage_embeddings.T
    )

    # BM25 negatives similarity: [B, K]
    bm25_logits = torch.einsum(
        "bd,bkd->bk",
        query_embeddings,
        bm25negatives
    )

    # Concatenate: [B, B+K]
    logits = torch.cat(
        [inbatch_logits, bm25_logits],
        dim=1
    ) / temperature

    # 5. Targets: diagonal = positives
    labels = torch.arange(B, device=logits.device)

    # 6. Cross-entropy
    loss = F.cross_entropy(logits, labels)

    return loss
```

```python
def compute_ir_metrics_ranx_TRAIN(query_embs, doc_embs, references, k=10):
    scores = (query_embs @ doc_embs.T)

    topk_values, topk_indices = torch.topk(scores, k=k, dim=1)

    topk_values = topk_values.cpu().numpy()
    topk_indices = topk_indices.cpu().numpy()
    qrels_dict = {str(i): {str(ref): 1} for i, ref in enumerate(references)}

    # run: { query_id: { doc_id: score } }
    run_dict = {}
    for i in range(len(topk_indices)):
        run_dict[str(i)] = {
            str(topk_indices[i][j]): float(topk_values[i][j])
            for j in range(k)
        }

    qrels = Qrels(qrels_dict)
    run = Run(run_dict)

    # 4. Расчет средних метрик
    metrics = [f"mrr@{k}"]

    res = evaluate(
        qrels,         # ground truth
        run,
        metrics,
        return_std=True,        # сохраняем std
        save_results_in_run=True,
    )

    # Получаем оценки для каждого отдельного запроса

    scores_mrr = list(run.scores[f"mrr@{k}"].values())

    return np.mean(scores_mrr)

    #print(f"Mean MRR@{k}: {np.mean(scores_mrr):.3f}, STD: {res_mrr.standard_error:.3f}")
```
